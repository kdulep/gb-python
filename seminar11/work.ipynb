Даны две функции:
f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
Требуется:
1. Найти точки пересечения этих функций.
2. Построить графики функций в одной системе координат
3. Построить график общей функции

```python
from sympy import symbols
from sympy import *
#import matplotlib.pyplot as plt
```

```python
x, y, t = symbols("x, y, t")
```

```python
fx=x**3-50*x
```

```python
gx=-x**4+88*x**2-241
```

```python
fx
```

```python
gx
```

```python
t=fx-gx
```

```python
t
```

```python
x1,x2,x3,x4=solve(t)
```

```python
solve(t)
```

```python
x1=x1.evalf(chop=True)
x2=x2.evalf(chop=True)
x3=x3.evalf(chop=True)
x4=x4.evalf(chop=True)
```

```python
x1, x2, x3, x4
```

```python
y1=fx.subs(x,x1)
y2=fx.subs(x,x2)
y3=fx.subs(x,x3)
y4=fx.subs(x,x4)
```

```python
y1, y2, y3, y4
```
